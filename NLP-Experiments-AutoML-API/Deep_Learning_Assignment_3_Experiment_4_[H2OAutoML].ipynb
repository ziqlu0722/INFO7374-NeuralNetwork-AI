### Combine results from 4 APIs

```
import pandas as pd
```

```
amazon = pd.read_csv('amazon_result.csv') 
google = pd.read_csv('google_result.csv') 
ibm = pd.read_csv('ibm_results.csv') 
microsoft = pd.read_csv('microsoft_result.csv') 
```

```
amazon.rename(columns={'score': 'score_amazon'}, inplace=True)
google.rename(columns={'score': 'score_google'}, inplace=True)
ibm.rename(columns={'score': 'score_ibm'}, inplace=True)
microsoft.rename(columns={'score': 'score_microsoft'}, inplace=True)
```

```
new1 = pd.merge(amazon,google,on="text",how='outer')
new2 = pd.merge(new1,ibm,on="text",how='outer')
new3 = pd.merge(new2,microsoft,on="text",how='outer')

new3.drop_duplicates(subset ="text", 
                     keep = False, inplace = True) 
final = new3[['score_amazon', 'score_google','score_microsoft', 'score_ibm','label']]
```

```
final
```

```
final.fillna(0)
```

```
final.to_csv("h2o_data.csv",index=False)
```

## Running H2O model

```
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```
dataset = h2o.import_file("h2o_data.csv")

train,test,valid= dataset.split_frame(ratios=[.7, .15])
```

```
aml = H2OAutoML(max_models=20, seed=1,max_runtime_secs =300)
```

```
x = train.columns
y = "label"
x.remove(y)




train[y] = train[y].asfactor()
aml.train(x=x, y=y, training_frame=train)
```

AutoML Results

```
lb = aml.leaderboard
print (lb)
```

```
test_y = aml.leader.predict(test)
```

Predictions

```
test_y['predict']
test['label']
```

```
#test_y.export_file("preds.csv") 
h2o.export_file(test_y,"preds.csv") 
h2o.export_file(test,"test_res.csv") 

```

```
import numpy as np
preds_res = pd.read_csv('preds.csv') 
test_res = pd.read_csv('test_res.csv') 
test_res=test_res.fillna(0)
preds_res_1=preds_res['predict'].astype(float)
from sklearn.metrics import accuracy_score

print("H2O auto ML accuracy score is "+str(accuracy_score(preds_res_1,test_res['label'])))
```
